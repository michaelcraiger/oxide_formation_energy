```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr
import numpy as np
```

```python
ys = [
    -0.29,
    -0.29,
    -0.13, 
    -0.06, 
    -0.02, 
    0.04
]
xs = [
    (0.67)*(-0.32)+(0.33)*(-0.25),
    0.8*((0.5)*(-0.32)+(0.5)*(-0.13)),
    (0.8)*((0.5)*(-0.32)+(0.5)*(-0.05)),
    (0.6)*((0.5)*(-0.32)+(0.5)*(-0.14)),
    (0.4)*((0.33)*(-0.32)+(0.67)*(-0.1)),
    (0)*((0.67)*(-0.32)+(0.33)*(-0.25)),

]

eles =[
    "W",
    "Nb", 
    "Mo",
    "Ir",
    "V", 
    "Mn"
]
```

```python
ele2col = {
    "V": "#ffcccb",
    "Mn": "purple",
    "Nb": "cyan",
    "Mo": "#FFA500",
    "W": "fuchsia",
    "Ir": "b",
} 
plt.plot([0, -0.3], [0, -0.3], linestyle='--', label='$y=x$')
xs_ = []
ys_ = []
for idx, ele in enumerate(eles):
    xs_.append(xs[idx])
    ys_.append(ys[idx])
    plt.plot(xs[idx], ys[idx], 'o', color=ele2col[ele],markersize=9, markeredgecolor='k',label=ele)
plt.ylabel("$\delta_{Cr_{1-z}B_{z}O_{2}}$ (eV/atom)", fontsize=14)
plt.xlabel("$\sigma_{Cr_{1-z}B_{z}O_{2}}$ (eV/atom)", fontsize=14)
sns.regplot(xs_, ys_, label='$y=1.24x+0.05$')
plt.legend(fontsize=10)
plt.ylim(-0.35, 0.2)
plt.xlim()
```

```python
pearsonr(xs_, ys_)
```

```python
np.polyfit(xs_, ys_, 1) 
```
